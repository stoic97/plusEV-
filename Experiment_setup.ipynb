```python
import mlflow
import os

# Configure MLflow
os.environ['MLFLOW_TRACKING_URI'] = 'mlruns'  # Local tracking
mlflow.set_tracking_uri('mlruns')

# Create experiment
experiment_name = "trading_model_development"
mlflow.set_experiment(experiment_name)

# Test tracking setup
with mlflow.start_run(run_name="setup_test"):
    mlflow.log_param("test_param", "success")
    mlflow.log_metric("test_metric", 1.0)
```
